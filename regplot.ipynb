```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
diamonds = diamonds.sample(n=250, random_state = 250)
```

### 1: Basic Regplot with diamonds! (h2)

```python
sns.set_style('dark')
```

```python
sns.regplot(x="carat", y="price", data=diamonds,
           marker ='d');
```

### Taking the scatter out, leaving regression line only

```python
sns.regplot(x="carat", y="price", data=diamonds, scatter=False);
#will need limit 
plt.gca().set(xlim=(0,2.6))
```

### 2:Confidence Interval

```python
sns.regplot(x="carat", y="price", data=diamonds, ci=False);
```

### 4: Order for a better fit line 

```python
sns.regplot(x="carat", y="price", data=diamonds, order=2);
```

### 5: Robust to ignore outlier(s)

```python
sns.regplot(x="carat", y="price", data=diamonds, robust=True);
```

### 6:Lowess

```python
sns.regplot(data=diamonds, x="carat", y="price", lowess=True)
```

### 7: Styling Graphs

```python
sns.regplot(x="carat", y="price", data=diamonds, 
        scatter_kws={"color": "purple", "s": 20, "alpha": 0.5},
        line_kws={'lw': '4',"color": "red", 'linestyle': '-.'})
```
